```python
# Write a program that uses input to prompt a user for their name and then welcomes them.
name = input('Enter Your Name : ')
print("Hello ", name)
```

```python
# Write a program to prompt the user for hours and rate per hour to compute gross pay.
hours = int(input("Enter Hours : "))
rate = int(input("Enter Rate : "))
print("Pay : ", hours*rate)
```

```python
# conversion from Celsius to Farenheit
tempInCel = float(input("Enter Temperature in Celsius : "))
tempInFar = (9/5)*tempInCel + 32
print(f"Temperature in Farenheit : {tempInFar}")
```

```python
# calculation of area, diagonal, and perimeter of a rectangle
from math import sqrt
height = float(input("Enter Height : "))
width = float(input("Enter Width : "))
area = height*width
diagonal = sqrt((height**2) + (width**2))
perimeter = 2*(height+width)
print(f"Area : {area}")
print(f"Diagonal : {diagonal}")
print(f"Perimeter : {perimeter}")
```

```python
# calculation of area of a trapezoid
height = float(input("Enter Height : "))
base1 = float(input("Enter Base-1 : "))
base2 = float(input("Enter Base-2 : "))
area = ((base1 + base2)/2) * height
print("Area : ", area)
```

```python
# calculation of surface volume and area of a cylinder
from math import sqrt, pi
height = float(input("Enter Height : "))
radius = float(input("Enter Radius : "))
area = 2*pi*radius*height + 2*pi*(radius**2)
volume = pi*(radius**2)*height
print(f"Volume of the Cylinder is {volume}")
print(f"Area of the Cylinder is {area}")
```

```python
### calculation of arc length of an angle and sector area of a circle
from math import pi
radius = float(input("Enter Radius of the Circle : "))
angle = float(input("Enter Angle of the Arc in Degree : "))
angleToRadian = (angle*pi)/180
arcLen = radius*angleToRadian
area = (angle/360)*pi*(radius**2)
print(f"Arc Length of the angle is {arcLen}")
print(f"Sector Area of the circle is {area}")
```
